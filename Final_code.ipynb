```python
from google.colab import drive
import pandas as pd

# Google Drive 마운트
# drive.mount('/content/drive')

# CSV 파일 경로
file_path = '/content/drive/My Drive/AI_Contest/melon_gerne_top50_bpm_tokenized.csv'

# 데이터프레임 로드
# encoding='euc-kr' 추가하여 파일 인코딩을 명시
df = pd.read_csv(file_path, encoding='euc-kr')

# 전체 데이터에서 BPM 최댓값과 최솟값 계산
max_bpm = df['BPM'].max()
min_bpm = df['BPM'].min()
print(max_bpm)
print(min_bpm)

# BPM 범위를 3등분
bpm_range = max_bpm - min_bpm
interval = bpm_range / 3

# 각 구간의 경계 계산
boundary1 = min_bpm + interval
boundary2 = min_bpm + 2 * interval

# 결과 확인
print(f"BPM Range: {min_bpm} - {max_bpm}")
print(f"Interval Size: {interval}")
print(f"Boundary 1: {boundary1}")
print(f"Boundary 2: {boundary2}")

# 사용자 입력 받기
user_bpm_preference = input("선호하는 BPM 속도를 입력하세요 (빠름, 보통, 느림): ").strip()

# BPM 필터링 함수 정의
def filter_songs_by_bpm_function(preference, data):
    if preference == '빠름':
        return data[data['BPM'] > 155]
    elif preference == '보통':
        return data[(data['BPM'] >= 102) & (data['BPM'] <= 155)]
    elif preference == '느림':
        return data[data['BPM'] < 102]
    else:
        return pd.DataFrame()  # 빈 데이터프레임 반환

# 사용자 선호 BPM에 맞는 음악 필터링
filtered_songs_by_bpm = filter_songs_by_bpm_function(user_bpm_preference, data)

# 필터링된 음악 제목 출력
if not filtered_songs_by_bpm.empty:
    print("추천 음악 제목:")
    print(filtered_songs_by_bpm.head())
else:
    print("해당 BPM 속도의 음악이 없습니다.")

# 예시: '빠름' BPM에 맞는 음악 필터링
# filtered_songs = filter_songs_by_bpm('빠름', df)
# print(filtered_songs.head())
```

```python
## 가중치에 따른 점수 반영을 통한 랜덤 플레이리스트 생성
# 점수 반영 > 점수가 높을수록 플레이리스트에 반영될 확률이 높음.

import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 추천 음악 추출: 감정 기반으로 음악 필터링

emo_data = pd.read_csv('/content/drive/MyDrive/AI_Contest/predictions_output_with_emotions.csv')
# 'user_emo'가 예를 들어 '기쁨'이라는 감정이라면, 해당 감정에 맞는 데이터를 필터링
emotion_based_songs = emo_data[emo_data['Predicted_Emotion'] == user_emo]

# 사용자 선호 BPM에 맞는 음악 필터링
filtered_songs_by_bpm = filter_songs_by_bpm_function(user_bpm_preference, data)

# 감정 기반 음악과 BPM 기반 음악 가중치 적용 (7:3 비율)
w1, w2 = 0.7, 0.3

# 감정 기반 곡과 BPM 기반 곡을 같은 데이터 프레임에서 필터링 했다고 가정
emo_data = pd.DataFrame()
emo_data['Emotion_Score'] = np.where(emotion_based_songs['Predicted_Emotion'] == user_emo, 1, 0) * w1
emo_data['BPM_Score'] = filtered_songs_by_bpm['BPM'].apply(lambda bpm: w2 if bpm == user_bpm_preference else 0)


##########
# Emotion_Score 추가 (감정 기반 곡은 emotion_weight만큼의 점수 부여)
emotion_based_songs['Emotion_Score'] = np.where(emotion_based_songs['Predicted_Emotion'] == user_emo, 1, 0) * w1
emotion_based_songs['BPM_Score'] = 0  # BPM 관련 점수는 없으므로 0으로 설정

filtered_songs_by_bpm = filtered_songs_by_bpm.copy()

# BPM_Score 추가 (BPM 기반 곡은 bpm_weight만큼의 점수 부여)
filtered_songs_by_bpm['Emotion_Score'] = 0  # 감정 관련 점수는 없으므로 0으로 설정
filtered_songs_by_bpm['BPM_Score'] = w2

# 두 데이터프레임 병합 (중복된 곡을 제거하면서 결합)
merged_songs = pd.concat([emotion_based_songs, filtered_songs_by_bpm]).drop_duplicates(subset=['Title'])

# Total_Score 계산 (Emotion_Score + BPM_Score)
merged_songs['Total_Score'] = merged_songs['Emotion_Score'] + merged_songs['BPM_Score']

# # 점수를 기반으로 5곡 랜덤 선택
# selected_songs = random.choices(merged_songs['Title'].tolist(), weights=merged_songs['Total_Score'].tolist(), k=5)

# # 랜덤 플레이리스트 출력
# print("랜덤 플레이리스트 (5곡):", selected_songs)

# 곡의 Title, Artist, Total_Score를 포함한 5곡을 가중치 기반으로 랜덤 선택
selected_songs = merged_songs.sample(n=5, weights='Total_Score', replace=False)

# 선택된 곡의 Title, Artist, Total_Score 출력
print("랜덤으로 선택된 곡 목록:")
for _, song in selected_songs.iterrows():
    print(f"{song['Title']} - {song['Artist']}, Total Score: {song['Total_Score']}")
```
